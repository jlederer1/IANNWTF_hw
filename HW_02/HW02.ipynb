```python
import numpy as np
np.random.seed(42)

n = 100 
x = np.random.rand(n)
x[:5];
```

```python
n_range = np.linspace(-0.5, 1, n)
def f(x): return x**2 - x**3

t = f(n_range)
t[:5];
```

https://matplotlib.org/stable/users/explain/api_interfaces.html#api-interfaces  

```python
import matplotlib.pyplot as plt

plt.subplot(1, 3, 1)
plt.scatter(x, np.zeros(n), s=10, c='b', marker="x")

plt.subplot(1, 3, 2)
plt.scatter(t, np.zeros(n), s=10, c='r', marker="o")

plt.subplot(1, 3, 3)
plt.scatter(n_range, t, c='m', marker=".")
plt.show()

# print("\n X: ", n_range[:5], "\n Y: ", t[:5])
```
