## A2.1

```python
n = 100 
n_range = np.linspace(-0.5, 1, n)
```

```python
np.random.seed(42)

x = np.random.rand(n)
x[:5]
```

```python
t = np.zeros(n)
for i in range(n):
  t[i] = x[i]**3 - x[i]**2
t[:5]
```

https://matplotlib.org/stable/users/explain/api_interfaces.html#api-interfaces  

```python
import matplotlib.pyplot as plt

def f(x): return x**3 - x**2
t2 = f(n_range)
t2[:5]
```

```python
plt.subplot(1, 3, 1)
plt.scatter(x, np.zeros(n), s=10, c='b', marker="x")

plt.subplot(1, 3, 2)
plt.scatter(t, np.zeros(n), s=10, c='r', marker="o")

plt.subplot(1, 3, 3)
plt.scatter(n_range, t2, c='m', marker=".")
plt.show()
print("\n X: ", n_range[:5], "\n Y: ", t2[:5])
```

https://matplotlib.org/stable/gallery/color/named_colors.html 

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.scatter(x, np.zeros(n), s=10, c='b', marker="x", label='Random numbers x [0,1)')
ax1.scatter(t, np.zeros(n), s=10, c='r', marker="o", label='Target numbers f(x)')

ax2.plot(n_range, t2, c='m', label='Target function f')

plt.legend(loc='upper left')
plt.show()
```
